```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df=df.iloc[:,2:]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.1,random_state=1)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
scalar.fit(X_train)
X_train_scalar=scalar.transform(X_train)
X_test_scalar=scalar.transform(X_test)
```

```python
X_train
```

```python
X_train_scalar
```

```python
X_test_scalar=pd.DataFrame(X_test_scalar,columns=X_train.columns)
X_train_scalar=pd.DataFrame(X_train_scalar,columns=X_test.columns)
```

```python
X_test_scalar
```

```python
X_train_scalar
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,6))
ax1.scatter(X_train['Age'],X_train['EstimatedSalary'])
ax1.set_title('Before Scaling')
ax2.scatter(X_train_scalar['Age'],X_train_scalar['EstimatedSalary'])
ax2.set_title('After Scaling')
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,6))
# Pass the DataFrame and specify 'x' and 'y' columns for kdeplot
sns.kdeplot(x=X_train['Age'], y=X_train['EstimatedSalary'], ax=ax1) # Specify ax=ax1 to plot on the first subplot
ax1.set_title('Before Scaling')
sns.kdeplot(x=X_train_scalar['Age'], y=X_train_scalar['EstimatedSalary'], ax=ax2) # Specify ax=ax2 to plot on the second subplot
ax2.set_title('After Scaling')
```
